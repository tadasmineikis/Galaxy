The files should be generated by pegase or downloaded from padova cmd http://stev.oapd.inaf.it/cgi-bin/cmd and edited if necessary to resamble file above.

The main features the file must have:
- first line with the column names, and filter names in file must mach filter names in the dictionary PASSBANDS. 

NOTE: in the example above the metallicity column is obsolete and not necessary.

```python
PASSBANDS=['U', 'B', 'V', 'R', 'I', 'J', 'H', 'K']
#PASSBANDS=['M','MWD','MBHNS','FUV','NUV','LHa','2MASSKs','2MASSJ','2MASSH','U','B','V','Ic','Rc','gi','SDSSu','SDSSg','SDSSr','SDSSi','SDSSz']
```

```python
DATA={}
Z_LIST=[]
fobj=open(DIR+'list','r')
for line in fobj:
    z=line.split()[0]
    name=line.split()[1]
    Z_LIST.append(z)
    DATA[z]=genfromtxt(DIR+name,names=True)
    print name, z
```

Load the file with ages

```python
AGES=genfromtxt(DIR+'ages.dat')
```

Prepare interpolation functions

```python
from scipy.interpolate import UnivariateSpline as spline
PASSB={}
for passb in PASSBANDS:
    PASSB[passb]=zeros((AGES.size,len(Z_LIST)))
    for idx, z in zip(range(len(Z_LIST)),Z_LIST):
        f=spline(log10(DATA[z]['Age'])-6,DATA[z][passb],k=1,s=0)
        PASSB[passb][1:,idx]=f(log10(AGES[1:]))
        PASSB[passb][0,idx]=f(log10(5))
```

Check the interpolation, results are saved as images in the specified DIR directory

```python
plt.figure(figsize=(19,12), dpi=100)
for passb in PASSBANDS:
    for idx, z in zip(range(len(Z_LIST)),Z_LIST):
        plot(DATA[z]['Age']/1e6, DATA[z][passb]+idx)
        plot(AGES, PASSB[passb][:,idx]+idx,'ok',markersize=2)
        xlabel('time, Myr', size=30)
        ylabel(passb,size=30)
    xscale('log')
    if passb =='LHa':
        yscale('log')
    plt.savefig(DIR+passb+".png", dpi=300)
    plt.cla()
plt.close()
```

Write out the files

```python
for passb in PASSBANDS:
    if passb == 'LHa':
        savetxt(DIR+passb+"_z8_a2600",PASSB[passb]/1e30,fmt="%.4e")
    elif passb =='M' or passb == 'MWD' or passb == 'MBHNS':
        savetxt(DIR+passb+"_z8_a2600",PASSB['MBHNS']+PASSB['MWD']+PASSB['M'],fmt="%.4e")
    else:
        savetxt(DIR+passb+"_z8_a2600",pow(10,-0.4*PASSB[passb]),fmt="%.4e")
```
